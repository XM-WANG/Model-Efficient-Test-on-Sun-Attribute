```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
'''案例1'''
x = np.linspace(-1,1,50)#-1到1平分取50个点
y = x**2
print(type(y))
plt.plot(y)#画点
plt.show()
```

```python
'''多图片+单图片多线+线参数'''
x = np.linspace(-1,1,50)
y1 = x**2
y2 = 2*x+1

#plt.figure(num=3, figsize=(8,5))
plt.figure()#打印多个图片的时候，figure下接的代码块是本张图片
plt.plot(x,y1)
plt.show()

plt.figure()
plt.plot(x,y1)
plt.plot(x,y2,color = 'red',linewidth=3.0, linestyle='--')#参数使用
plt.show()
```

```python
'''坐标轴描述+坐标轴标尺'''
x = np.linspace(-1,2,50)
y1 = x**2
y2 = 2*x+1

plt.figure()

plt.xlim((-1,2 ))#x轴取值范围
plt.ylim((-2,3))#y轴取值范围

plt.xlabel(("I am x"))#x轴描述
plt.ylabel(("I am y"))#y轴描述

new_ticks = np.linspace(-1,2,5)
plt.xticks(new_ticks)
plt.yticks([-2,-1.8,-1,1.22,3],[r"$really\ bad$", "bad", "normal", "good","really good"])

plt.plot(x,y1)
plt.plot(x,y2,color = 'red',linewidth=1.0, linestyle='--')#参数使用
plt.show()
```

```python
'''坐标轴位置'''
x = np.linspace(-1,2,50)
y1 = x**2
y2 = 2*x+1

plt.plot(x,y1)
plt.plot(x,y2,color = 'red',linewidth=1.0, linestyle='--')

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')#确定那两条边为坐标轴
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data',0))#确定坐标轴的原点位置
ax.spines['left'].set_position(('data',0))

plt.show()
```

```python
'''图例设置'''
x = np.linspace(-1,2,50)
y1 = x**2
y2 = 2*x+1

plt.plot(x,y1,label="cc")
plt.plot(x,y2,color = 'red',linewidth=1.0, linestyle='--')

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

plt.legend(labels = ['aa','bb'],loc = 'best')
#handle:画哪些线，要用变量接受plot的返回; label:列表，采用列表里的名字取代原来的label

plt.show()
```

```python
'''设置注解'''
x = np.linspace(-3,3,50)
y = 2*x + 1

plt.figure(num = 1, figsize = (8,5),)
plt.plot(x,y,)
#plt.scatter(x,y) #散点图画法

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

x0 = 1
y0 = 2*x0 + 1
plt.scatter(x0,y0,s=50, color='b')#画点
plt.plot([x0,x0],[y0,0],'k--',lw=2.5)
#两点确定一条线，第一个list记录两点的x坐标，第二个list记录两点的y坐标，k表示颜色，--线的样子，lw线宽

plt.annotate(r'$2x+1=%s$'%y0,xy=(x0,y0),xycoords = 'data',xytext=(+30,-30),textcoords='offset points',fontsize = 16, arrowprops = dict(arrowstyle = '->',connectionstyle='arc3,rad=.2'))
'''
参数说明：
1. $2x+1=%s$'%y0 ： 要打印的内容
2. xy=(x0,y0)    : 要打印内容的位置，坐标
3. xycoords      ： 基于坐标系
4. xytext        ： 打印内容的偏移量
5. xycoods       ： 基于点的偏移
'''
plt.show()
```
