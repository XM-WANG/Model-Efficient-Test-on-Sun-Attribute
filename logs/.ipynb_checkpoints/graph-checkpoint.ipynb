```python
import matplotlib.pyplot as plt
import numpy as np
import re
```

- 字符串匹配算法

```python
def find_str(pattern, string):
    start_index = 0
    end_index = 0
    index_list = []
    while True:
        result = re.search(pattern,string[end_index:])
        if result is None:
            break
        else:
            start_index, end_index = end_index+result.start(), end_index+result.end()
            index_list.append([start_index,end_index])
    return index_list
```

- 向下采样，loss值过多，画出图像波动过大，下采样后有更好的视觉效果

```python
def down_sampling(high_list,rate):
    low_list=[]
    for i in range(len(high_list)):
        if i%rate == 0:
            low_list.append(high_list[i])
    return low_list  
```

- VGGNet 和 ResNet Epoch对Recall的影响对比

```python
with open("part1/train_log1.txt","r") as f:
    vgg_log = f.read()
vgg_index = find_str("Top 5 Recall",vgg_log)
del vgg_index[0]
del vgg_index[len(vgg_index)-1]
vgg_recall = []
for i in vgg_index:
    recall = float(vgg_log[i[0]+13:i[1]+6])
    vgg_recall.append(recall/100)

with open("part1/train_log2.txt","r",encoding='UTF-8') as f:
    res_log = f.read()
res_index = find_str("Top 5 Recall",res_log)
del res_index[0]
del res_index[len(res_index)-1]
res_recall = []
for i in res_index:
    recall = float(res_log[i[0]+13:i[1]+6])
    res_recall.append(recall/100)
```

- 画图

```python
x = np.linspace(0,100,100)
vgg_y = np.asarray(vgg_recall)
res_y = np.asarray(res_recall)
plt.xlabel(("Epoch"))#x轴描述
plt.ylabel(("Recall"))#y轴描述
plt.plot(x,vgg_y,label="VGG-NET",color = 'darkorange',linewidth=1.0)
plt.plot(x,res_y,label="RES-NET",color = 'darkblue',linewidth=1.0)
plt.legend(labels = ['VGG-NET','RES-NET'],loc = 'best')
plt.savefig('recall.jpg',dpi=150)
plt.show()
```
